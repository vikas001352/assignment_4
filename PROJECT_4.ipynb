```python
ANS.1


Class:
A class is like a blueprint or a set of instructions that defines how objects of that class will be created and what attributes (data members) and behaviors (methods) they will have. It encapsulates data and behavior into a single unit, making it easy to organize and manage code.

Object:
An object is an instance of a class. When you create an object, it becomes a specific instance that adheres to the class's structure and behavior. Objects can have their own unique data, but they share the same behavior as defined by the class.

Example:
Let's consider an example of a class Car, which represents the blueprint for creating car objects. The class will have data members like make, model, year, and methods like start(), stop(), and accelerate().


class Car:
    def __init__(self, make, model, year):
        self.make = make
        self.model = model
        self.year = year
        self.is_running = False

    def start(self):
        self.is_running = True
        print(f"{self.make} {self.model} started.")

    def stop(self):
        self.is_running = False
        print(f"{self.make} {self.model} stopped.")

    def accelerate(self, speed):
        if self.is_running:
            print(f"{self.make} {self.model} accelerating to {speed} km/h.")
        else:
            print(f"Please start the {self.make} {self.model} before accelerating.")

# Create objects of the Car class
car1 = Car("Toyota", "Corolla", 2022)
car2 = Car("Honda", "Civic", 2021)

# Using the methods of the Car objects
car1.start()
car1.accelerate(50)
car1.stop()

car2.start()
car2.accelerate(60)
car2.stop()


OUTPUT.

Toyota Corolla started.
Toyota Corolla accelerating to 50 km/h.
Toyota Corolla stopped.
Honda Civic started.
Honda Civic accelerating to 60 km/h.
Honda Civic stopped.


ANS.2

1=Encapsulation: Encapsulation is the process of bundling data (attributes) and methods (functions) that operate on the data within a single unit called a class. The class acts as a protective container, and the internal details are hidden from the outside world. This allows data to be accessed and manipulated only through the defined methods, ensuring data integrity and security.
2=Abstraction: Abstraction is the process of simplifying complex systems by breaking them down into smaller, manageable units. In OOP, abstraction allows you to create abstract classes or interfaces that provide a blueprint for subclasses. Abstract classes contain abstract methods that define the interface without providing an implementation. Subclasses must implement these abstract methods, allowing for polymorphism and code reuse.
3=Inheritance: Inheritance is a mechanism that allows a class (subclass) to inherit the properties and behaviors of another class (superclass). It establishes a parent-child relationship between classes, where the subclass inherits attributes and methods from the superclass. Inheritance promotes code reusability, as subclasses can extend and modify the functionality of the superclass.

Polymorphism: Polymorphism is the ability of a class to take on multiple forms or to exhibit multiple behaviors. It allows objects of different classes to be treated as objects of a common superclass. Polymorphism is achieved through method overriding (when a subclass provides a specific implementation for an inherited method) and method overloading (when a class has multiple methods with the same name but different parameters). Polymorphism enables more flexible and generic coding practices, enhancing the flexibility of the code.


ANS.3

class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def display_info(self):
        print(f"Name: {self.name}, Age: {self.age}, Grade: {self.grade}")

# Create instances of the Student class
student1 = Student("Alice", 15, "10th")
student2 = Student("Bob", 16, "11th")

# Call the display_info() method for each student
student1.display_info()
student2.display_info()



ANS.4

In Object-Oriented Programming (OOP), self is a special parameter that is used as a reference to the current instance of the class. It is a convention in Python (and some other object-oriented languages) to use the name self as the first parameter in the methods of a class.

When you call a method on an object, Python automatically passes the reference to the object as the first argument to the method. This reference is then assigned to the self parameter, allowing the method to access and manipulate the object's attributes and methods.

Using self is essential for properly handling the instance-specific data within a class. It allows methods to operate on the attributes and methods of the specific instance on which the method is called.

Example:
    class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def bark(self):
        print(f"{self.name} is barking!")

# Creating two Dog objects
dog1 = Dog("Buddy", 3)
dog2 = Dog("Max", 5)

# Calling the bark() method on each object
dog1.bark()  # Output: Buddy is barking!
dog2.bark()  # Output: Max is barking!


ANS.5

ngle Inheritance:
Single inheritance occurs when a subclass inherits from a single superclass. It is the most common type of inheritance.
Example:
    # Base class
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        pass

# Derived class
class Dog(Animal):
    def __init__(self, name):
        super().__init__("Dog")
        self.name = name

    def make_sound(self):
        return "Woof!"

# Create an instance of Dog and call methods
dog1 = Dog("Buddy")
print(dog1.species)   # Output: Dog
print(dog1.make_sound())   # Output: Woof!

Multiple Inheritance:
Multiple inheritance occurs when a subclass inherits from two or more superclasses. In Python, a subclass can inherit from multiple classes by listing them within parentheses.
Example:
    # Base classes
class A:
    def method_a(self):
        return "Method A"

class B:
    def method_b(self):
        return "Method B"

# Derived class with multiple inheritance
class C(A, B):
    def method_c(self):
        return "Method C"

# Create an instance of C and call methods
c_instance = C()
print(c_instance.method_a())   # Output: Method A
print(c_instance.method_b())   # Output: Method B
print(c_instance.method_c())   # Output: Method C

Hierarchical Inheritance:
Hierarchical inheritance occurs when multiple subclasses inherit from the same superclass.
Example:
    # Base class
class Vehicle:
    def __init__(self, vehicle_type):
        self.vehicle_type = vehicle_type

# Subclasses inheriting from Vehicle
class Car(Vehicle):
    def __init__(self, make):
        super().__init__("Car")
        self.make = make

class Motorcycle(Vehicle):
    def __init__(self, make):
        super().__init__("Motorcycle")
        self.make = make

# Create instances of Car and Motorcycle and call methods
car1 = Car("Toyota")
motorcycle1 = Motorcycle("Harley Davidson")
print(car1.vehicle_type)   # Output: Car
print(motorcycle1.vehicle_type)   # Output: Motorcycle


In each example, we can observe the inheritance relationships between classes, where the subclass inherits attributes and methods from the superclass. This allows code reusability and promotes the creation of logical and hierarchical class structures.
```
